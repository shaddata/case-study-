```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to the python environment

```python
cars = pd.read_csv('cars_data.csv')
```

# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(10,6))
sns.countplot(x=cars['Buyer Gender'])
plt.grid(linestyle="--")
plt.show()
```

# 4. Find the top 5 cars based on their sales price

```python
top5sales = cars.sort_values(by='Sale Price',ascending=False).iloc[:5]
top5sales
```

# 5. Find the least 5 cars based on their Resell price

```python
least5cars = cars.sort_values(by='Resell Price').iloc[:5]
least5cars
```
